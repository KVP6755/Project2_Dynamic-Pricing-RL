```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Simulation settings
days = 30
initial_inventory = 100
base_price = 5000

inventory = initial_inventory

results = []
```

```python
for day in range(days):

    days_left = days - day

    # Fixed Pricing
    fixed_price = base_price

    # Time-Based Discount Pricing
    if days_left > 20:
        discount_price = base_price
    elif days_left > 10:
        discount_price = base_price * 0.9
    else:
        discount_price = base_price * 0.8

         # Simulated demand
    demand = np.random.randint(1, 8)

    bookings = min(demand, inventory)

    revenue_fixed = bookings * fixed_price
    revenue_discount = bookings * discount_price

    inventory -= bookings

    results.append([
        day + 1,
        days_left,
        bookings,
        fixed_price,
        discount_price,
        revenue_fixed,
        revenue_discount,
        inventory
    ])

    if inventory == 0:
        break

df = pd.DataFrame(results, columns=[
    "Day",
    "Days Left",
    "Bookings",
    "Fixed Price",
    "Discount Price",
    "Fixed Revenue",
    "Discount Revenue",
    "Remaining Inventory"
])

print(df)

print("\nTotal Fixed Revenue:", df["Fixed Revenue"].sum())
print("Total Discount Revenue:", df["Discount Revenue"].sum())

plt.plot(df["Day"], df["Fixed Revenue"], label="Fixed Pricing")
plt.plot(df["Day"], df["Discount Revenue"], label="Discount Pricing")
plt.xlabel("Day")
plt.ylabel("Revenue")
plt.title("Revenue Comparison")
plt.legend()
plt.show()
```
